```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#plotting the data
plt.plot(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```

```python
#Create the plot and annotate graph
plt.plot(x_axis, y_axis, marker ="*", linewidth=2, label="Boston", color="blue")
#Labels for the axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Set the window, or limit
plt.ylim(0, 45)
#Gridlines
plt.grid()
#Title
plt.title("PyBer Fare by Month")
#Legend
plt.legend()
```

```python
#Creating the figure
fig, ax = plt.subplots(figsize=(8, 4))
#Creating the plot
ax.plot(x_axis, y_axis, marker="*", color="green", linewidth=2, label="Boston")
#Labels for axes
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
#Window, or limit
ax.set_ylim(0, 45)
#Title
ax.set_title("PyBer Fare by Month")
#Legend
ax.legend()
#Gridlines
ax.grid()
```

```python
#creating the bar graph
plt.bar(x_axis, y_axis, color="green", label='Boston')
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.legend()
```

```python
#Creating the plot objects
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.invert_yaxis()
ax.legend()
```

```python
#Creating the figure
fig, ax = plt.subplots(figsize=(8, 4))
#Creating the plot
ax.plot(x_axis, y_axis, "o", marker="*", color="green", linewidth=2, label="Boston")
#Labels for axes
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
#Window, or limit
ax.set_ylim(0, 45)
#Title
ax.set_title("PyBer Fare by Month")
#Legend
ax.legend()
#Gridlines
ax.grid()
```

```python
#skill drill -- scatter plot
#creating it matlab approach
plt.plot(y_axis, x_axis, "o", color="red", label="Chicago")
plt.gca().invert_yaxis()
#labels
plt.ylabel("Date")
plt.xlabel("Fare($)")
#limit
plt.xlim(0, 45)
#title
plt.title("PyBer Fare by Month")
#legend
plt.legend()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#Utilizing list comprehension:
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
#skill drill object-oriented apporach, creating objects
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color="skyblue", label="Boston", s=[i*5 for i in y_axis], edgecolors='k', alpha=.8,
          linewidth=2)
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
ax.set_xlim(0, 50)
ax.invert_yaxis()
ax.set_title("PyBer Fare by Month")
ax.legend()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
#Curious about how to find the index needed if you have a large data set and didn't know the max
i = 0
j = len(y_axis)
ind = 0
for a in range(j):
    if y_axis[a] > i:
        i = y_axis[a]
    if y_axis[a] == i:
        ind = a
print(ind)
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

ax.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%',
    shadow=True,
    startangle=90,
    counterclock=False)

plt.show()
```
